```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import blankly
import tensorflow as tf
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
```

```python
# Load the data from blankly finance
exchange = blankly.CoinbasePro(portfolio_name="my cool portfolio")
paper_trade = blankly.PaperTrade(exchange, initial_account_values={'USD': 10000})
eth_hist = exchange.interface.history(symbol='ETH-USD', to='5y', resolution='6h', return_as='df')
eth_hist.at[594,'low'] = 305.29 # manually fixed the data as the stock crashed for a split second
```

```python
eth_data = eth_hist # so that we don't have to get data from blankly every time
# Used to plot OHLCV
def price_plot(price_type):  
    plt.figure(figsize=(15, 5))
    plt.plot(eth_data[price_type])
    plt.xticks(range(0, eth_data.shape[0], 1464), eth_data['time'].loc[::1464], rotation=0)
    if price_type == "volume":
        title = price_type.capitalize()
        plt.ylabel(title, fontsize=18)
    else:
        title = price_type.capitalize() + " Price"
        plt.ylabel('Price', fontsize=18)
    plt.title(title, fontsize=20)
    plt.legend([title], fontsize='x-large', loc='best')
    plt.show()
```

## Static Variables

```python
ndays_for_train_data = 64 # We can change
num_features = 5
```

-------------------


```python
# Plot OLHCV
price_plot("open")
price_plot("high")
price_plot("low")
price_plot("close")
price_plot("volume")
```

```python
'''Calculate percent change'''
eth_data['time'] = pd.to_datetime(eth_data['time'], unit='s')
eth_data['open'] = eth_data['open'].pct_change()
eth_data['high'] = eth_data['high'].pct_change()
eth_data['low'] = eth_data['low'].pct_change()
eth_data['close'] = eth_data['close'].pct_change()
eth_data['volume'] = eth_data['volume'].pct_change()
eth_data.dropna(how='any', axis=0, inplace=True) # Drop invalid data
```

```python
'''Normalize price data'''
min_price = min(eth_data[['open','high','low','close']].min(axis=0))
max_price = max(eth_data[['open', 'high', 'low', 'close']].max(axis=0))

eth_data['open'] = (eth_data['open'] - min_price) / (max_price - min_price)
eth_data['high'] = (eth_data['high'] - min_price) / (max_price - min_price)
eth_data['low'] = (eth_data['low'] - min_price) / (max_price - min_price)
eth_data['close'] = (eth_data['close'] - min_price) / (max_price - min_price)

'''Normalize volume data'''
min_volume = eth_data['volume'].min(axis=0)
max_volume = eth_data['volume'].max(axis=0)

eth_data['volume'] = (eth_data['volume'] - min_volume) / (max_volume - min_volume)
eth_data
```

```python
"""Create Traning, Validation, and Test split for Cross Validatoin"""

eth_data_indexes = sorted(eth_data.index.values) # index of the ETH data [1,2,3,..,N]
last_10p_index = sorted(eth_data.index.values)[-int(0.1*len(eth_data_indexes))] # Last 10% of the ETH dataset
last_20p_index = sorted(eth_data.index.values)[-int(0.2*len(eth_data_indexes))] # Last 20% of the ETH dataset

eth_data_train = eth_data[(eth_data.index < last_20p_index)]  # ETH Training data is 80% of total data
eth_data_val = eth_data[(eth_data.index >= last_20p_index) & (eth_data.index < last_10p_index)] # ETH Validation data is 10% of total data
eth_data_test = eth_data[(eth_data.index >= last_10p_index)] # ETH Test data is 10% of total data

# Convert pandas columns into arrays
train_data = eth_data_train.values
val_data = eth_data_val.values
test_data = eth_data_test.values
print('Training data shape: {}'.format(train_data.shape))
print('Validation data shape: {}'.format(val_data.shape))
print('Test data shape: {}'.format(test_data.shape))

eth_data_train.head()
```

```python
# Training data
X_train, y_train = [], []
for i in range(ndays_for_train_data, len(train_data)):
    X_train.append(train_data[i-ndays_for_train_data:i]) # Chunks of training data with a length of 64 days data
    y_train.append(train_data[:, 3][i]) #Value of Close Price of df-row (64+1)
X_train, y_train = np.array(X_train), np.array(y_train)

###############################################################################

# Validation data
X_val, y_val = [], []
for i in range(ndays_for_train_data, len(val_data)):
    X_val.append(val_data[i-ndays_for_train_data:i])
    y_val.append(val_data[:, 3][i])
X_val, y_val = np.array(X_val), np.array(y_val)

###############################################################################

# Test data
X_test, y_test = [], []
for i in range(ndays_for_train_data, len(test_data)):
    X_test.append(test_data[i-ndays_for_train_data:i])
    y_test.append(test_data[:, 3][i])    
X_test, y_test = np.array(X_test), np.array(y_test)

print(X_train.shape, y_train.shape)
print(X_val.shape, y_val.shape)
```

# Bi-LSTM Model

```python
def create_bilstm_model():
    in_seq = Input(shape = (ndays_for_train_data, num_features))
    
    layers = Bidirectional(LSTM(128, return_sequences=True))(in_seq)
    layers = Bidirectional(LSTM(128, return_sequences=True))(layers)
    layers = Bidirectional(LSTM(64, return_sequences=True))(layers)
    
    avg_pool = GlobalAveragePooling1D()(layers)
    max_pool = GlobalMaxPooling1D()(layers)
    
    conc = concatenate([avg_pool, max_pool])
    conc = Dense(64, activation='relu')(conc)
    output = Dense(1, activation='relu')(conc)
    
    model = Model(inputs=in_seq, outputs=output)
    model.compile(loss='mse', optimizer='adam', metrics=['mae','mape'])
    return model
```

```python
model = create_bilstm_model()
model.summary()
```
